```python
# BlackJack Version 5
import random 
import csv 

player_cards = []
computer_cards = []
pcards_sum = 0
comp_cards_sum = 0
acc_bal = 0

def getBet():
    x = 0
    if int(acc_bal) <= 0:
        print("Sorry, you do not have enough balance!")
        return x
    check_bet = False
    while (check_bet == False) :   
        try: 
            bet = int(input("Do you wish to place a bet for Rs. (21(all in)/50/100/200/500). For betting all your amount enter 21 "))
            print()
            if (bet != 50) and (bet != 100) and (bet != 200) and (bet != 500) and (bet !=21):
                print("Please choose a valid amount to bet")
                check_bet = False
            elif bet == 21:
                bet = acc_bal
                check_bet = True
            else: 
                check_bet = True
        except ValueError: 
            print("Invalid data type. Please enter a number")

    return bet

def getPlayerName():
    check_player_name = False
    while check_player_name == False: 
        try:
            name = input("Enter player name: ")
            if name.strip() == "":
                print("Please enter the player name")
            else:
                check_player_name = True
        except ValueError: 
            print("Invalid data type. Please enter a string value")
            
    return name

def drawCard(p_cards): 
    card_number = random.randint(1,13)
    
    if card_number == 1:
        p_cards.append(11)
        if getSum(p_cards) > 21:
            for i in range(0, len(p_cards)):
                if i == 11:
                    p_cards[i] = 1
                    break
                 
    elif card_number == 2:
        p_cards.append(2)
    elif card_number == 3:
        p_cards.append(3)
    elif card_number == 4:
        p_cards.append(4)
    elif card_number == 5:
        p_cards.append(5)
    elif card_number == 6:
        p_cards.append(6)
    elif card_number == 7:
        p_cards.append(7)
    elif card_number == 8:
        p_cards.append(8)
    elif card_number == 9:
        p_cards.append(9)
    elif card_number == 10: 
        p_cards.append(10)
    elif card_number == 11: 
        p_cards.append(11)
    elif card_number == 12: 
        p_cards.append(12)
    elif card_number == 13: 
        p_cards.append(13)

        
def getSum(p_cards):
    sum = 0
    for i in p_cards: 
        if (i == 11) or  (i == 12) or  (i == 13):
            sum += 10
        else:
            sum += i
    return sum

def checkSum(p_cards):
    sum = getSum(p_cards)
    if sum > 21:
        return False
    else:
        return True

def changeAceScore(p_cards):
    for i in range(0, len(player_cards)):
        if p_cards[i] == 11:
            p_cards[i] = 1
        
def displayCards(p_cards):
    for i in p_cards:
        if i == 1:
            print('''
              _________        
             |A        |         
             |+   *    |         
             |    !    |        
             |  *-+-*  |         
             |    |    |         
             |   ~~~  +|         
             |        V|         
              ~~~~~~~~~  
            ''')
        elif i == 2:
            print('''
              _________ 
             |2        | 
             |+        | 
             |    +    | 
             |         |
             |    +    |
             |        +|
             |        Z|
              ~~~~~~~~~ 
            ''')
        elif i == 3:
            print('''
              _________          
             |3        |        
             |+   +    | 
             |         |  
             |    +    | 
             |         |         
             |    +   +|       
             |        E| 
              ~~~~~~~~~  
            ''')
        elif i == 4:
            print('''
              _________          
             |4        |       
             |+        |       
             |  +   +  |      
             |         | 
             |  +   +  | 
             |        +|      
             |        b|        
              ~~~~~~~~~ 
            ''')

        elif i == 5:
            print('''
              _________ 
             |5        | 
             |+        | 
             |  +   +  |
             |    +    |           
             |  +   +  |         
             |        +|        
             |        S|     
              ~~~~~~~~~          
            ''')
        elif i == 6:
            print('''
              _________          
             |6        |         
             |+ +   +  |         
             |         |        
             |  +   +  |        
             |         | 
             |  +   + +| 
             |        9| 
              ~~~~~~~~~ 
            ''')
        elif i == 7:
            print('''
              _________          
             |7        |      
             |+ +   +  |         
             |    +    |         
             |  +   +  |       
             |         |        
             |  +   + +| 
             |        L| 
              ~~~~~~~~~ 
            ''')
        elif i == 8:
            print('''
              _________  
             |8 +   +  |       
             |+        | 
             |  +   +  |        
             |         |        
             |  +   +  |         
             |        +|         
             |  +   + 8|         
              ~~~~~~~~~         
            ''')
        elif i == 9:
            print('''
              _________           
             |9 +   +  |
             |+        | 
             |  +   +  |
             |    +    |  
             |  +   +  | 
             |        +|
             |  +   + 6| 
              ~~~~~~~~~          
            ''')
        elif i == 10:
            print('''
              _________           
             |10+   +  |         
             |+   +    |     
             |  +   +  | 
             |         | 
             |  +   +  |
             |    +   +|
             |  +   +0l|
              ~~~~~~~~~          
            ''')
        elif i == 11:
            print('''
              _________          
             |J /~~|_  | 
             |+ | o`,  |
             |  | -|   |
             | =~)+(_= | 
             |   |- |  |
             |  `.o | +| 
             |  ~|__/ P|
              ~~~~~~~~~  
            ''')
        elif i == 12:
            print('''
              _________          
             |Q |~~~|  |        
             |+ /o,o\  |
             |  \_-_/  |
             | _-~+_-~ | 
             |  /~-~\  | 
             |  \o`o/ +|
             |  |___| Q| 
              ~~~~~~~~~  
            ''')
        elif i == 13:
            print('''
              _________  
             |K |/|\|  | 
             |+ /o,o\  |
             |  \_-_/  | 
             | ~-_-~-_ | 
             |  /~-~\  |
             |  \o`o/ +|
             |  |\|/| X|
              ~~~~~~~~~   
            ''')

def getWinner(player, computer, p_cards, c_cards):
    if checkSum(p_cards) == False:
        return computer
    elif checkSum(c_cards) == False:
        return player
    elif getSum(p_cards) > getSum(c_cards):
        return player
    elif getSum(p_cards) < getSum(c_cards):
        return computer
    elif getSum(p_cards) == getSum(c_cards):
        return ""
            
def drawInitialCards():
    cards_dealt = 1
    while cards_dealt <= 2:
        drawCard(player_cards)  
        drawCard(computer_cards)

        if cards_dealt == 1:
            print("Computer has the first card: ")
            displayCards(computer_cards)
    
        cards_dealt += 1
    
    print(player_name, "you have the cards: ")
    displayCards(player_cards)
    

def checkWin(p1_cards, p2_cards, p1_name):
    win = False
    p1_sum = getSum(p1_cards)
    p2_sum = getSum(p2_cards)

    if p1_sum == 21 and p2_sum == 21:
        print("The game is a draw ")
        win = True
    elif p1_sum == 21:
        print(p1_name, "wins the game !")
        win = True
    elif p2_sum == 21:
        print("Computer wins the game !")
        win = True
        
    return win 
    
player_name = getPlayerName()

f = open("/Users/asmiirashi/Balance.csv", "r")
bal_read = csv.reader(f)
l1 = []

flag = 0
for i in bal_read:
    l1.append(i)
    if player_name == i[0]:
        acc_bal = int(i[1])
        print()
        print("Your account balance is Rs.", acc_bal)
        print()
        flag = 1
                    
if flag == 0: 
    acc_bal = 2500
    l1.append([player_name, str(acc_bal)])
    
if int(acc_bal) <= 50:
    acc_bal = 2500
    print()
    print("Your balance has been refilled and your current balance is Rs.", acc_bal)
    print()
        
f.close()


file = open("/Users/asmiirashi/Balance.csv", "w")
writer = csv.writer(file)
writer.writerows(l1)
file.close()

play_game = True
while play_game == True:
    total_bet = 0

    bet = getBet()
       
    total_bet += bet
    acc_bal -= bet
    
    print("Your current balance is", acc_bal)
    
    print()

    add_bet = True
    while add_bet == True: 
        c = input("Do you wish to bet more(yes/no) ?")
        if c.lower() == "yes":
            bet = getBet()
            total_bet += bet
            acc_bal -= bet
    
            print("Your current balance is", acc_bal)
            print()
            add_bet = True
            
        elif c.lower() == "no":
            add_bet = False
        else: 
            print("Please enter a valid choice(yes/no)")
            print()
            add_bet = True       

    
    player_cards = []
    computer_cards = []
    pcards_sum = 0
    comp_cards_sum = 0
    
    print("___________________________________________________________________________")
    print("Welcome to Asmii's BlackJack Game !! ")
    
    print()
    print("Now we will deal the cards")
    
    drawInitialCards()
    
    print("Your score is", getSum(player_cards))
    print()
    
    game = checkWin(player_cards, computer_cards, player_name )

    print("Hit- draw one more card")
    print("Stand - Stop drawing cards")
    print()
    
    while game == False:
        c = int(input("Do you wish to hit(0) / stand(1) ? "))
        if c == 0:
            drawCard(player_cards)
            x = 11
            if x in player_cards and getSum(player_cards) > 21:
                changeAceScore(player_cards)
    
            print("Now your cards are: ")
            displayCards(player_cards)
            print()
    
            print("Your score is", getSum(player_cards))
            print()

            if getSum(player_cards) == 21:
                break
            elif getSum(player_cards) > 21:
                game = True
  
        elif c == 1: 
            x = 11
            if x in computer_cards and getSum(computer_cards) > 21:
                changeAceScore(computer_cards)
    
            c_sum = getSum(computer_cards)
            
            
            while c_sum < 17:
                drawCard(computer_cards)
                c_sum = getSum(computer_cards)

            if getSum(computer_cards) == 21:
                break
                
            game = True


    winner = getWinner(player_name, "computer", player_cards, computer_cards)
    if winner == "":
        print("The computer's score is", getSum(computer_cards))
        print()
        print("The game is a draw")
    
    else:
        print("The computer's score is", getSum(computer_cards))
        print()
        print("The winner of the game is", winner, "!!") 

    if winner == player_name:
        acc_bal += total_bet*2

    print(player_name,"your current account balance is Rs.", acc_bal)
    print()
    
    f2 = open("/Users/asmiirashi/Balance.csv", "r")
    csv_reader = csv.reader(f2)
    temp = []
    for i in csv_reader:
        if player_name in i:
            i[1]= str(acc_bal)
        temp.append(i)
    f2.close()

    f3 = open("/Users/asmiirashi/Balance.csv", "w")
    csv_writer = csv.writer(f3)
    csv_writer.writerows(temp)
    f3.close()
    
    if acc_bal > 0:
        check_choice = False
        while check_choice == False:
            choice = input("Do you wish to play again ?(yes/no)")
            print()
            if choice.strip() == "":
                print("Please enter a choice(yes/no)")
            elif (choice.lower() != "yes") and (choice.lower() != "no"):
                print("Please enter a valid choice(yes/no)")
            else:  
                check_choice = True
                if choice.lower() == "yes":
                    play_game = True
                elif choice.lower() == "no":
                    play_game = False
    else:
        play_game = False

print()
print("Game Ended")        
            
        
```
